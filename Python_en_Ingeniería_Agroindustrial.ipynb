## 3 Ejemplos Prácticos y Sencillos

Aquí tienes tres fragmentos de código que ilustran tareas comunes en la carrera:

### 1. Cálculo del Índice de Cosecha

Este código ayuda a determinar si un lote está listo para ser procesado basándose en el peso del producto útil frente al peso total.

```
# Cálculo simple de eficiencia de cosecha
peso_fruto_util = 450.5  # kg
peso_biomasa_total = 600.0  # kg

indice_cosecha = (peso_fruto_util / peso_biomasa_total) * 100

print(f"El índice de cosecha es del {indice_cosecha:.2f}%")
```

### 2. Clasificador de Madurez (Lógica Básica)

Imagina un sensor que mide los Grados Brix (azúcar). Este código automatiza la decisión de a qué línea de producción enviar la fruta.

```
grados_brix = 12.5

if grados_brix < 10:
    clasificacion = "Inmaduro - Uso industrial (mermeladas)"
elif 10 <= grados_brix <= 14:
    clasificacion = "Óptimo - Consumo en fresco"
else:
    clasificacion = "Sobre-maduro - Fermentación/Alcohol"

print(f"Resultado: {clasificacion}")
```

### 3. Estimación de Pérdida por Humedad

En el secado de granos, es fundamental saber cuánta agua se ha perdido para evitar dañar el producto.

```
peso_inicial = 1000  # kg de grano húmedo
humedad_inicial = 0.25  # 25%
humedad_final = 0.12    # 12% (objetivo)

# Fórmula de balance de masa para sólidos constantes
peso_final = peso_inicial * ((1 - humedad_inicial) / (1 - humedad_final))

print(f"El peso final del lote tras el secado será: {peso_final:.2f} kg")
```
